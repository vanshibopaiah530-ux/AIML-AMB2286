```python
lower = 1
upper = 50

print(f"Prime numbers between {lower} and {upper} are:")

# Outer loop: Iterates through each number in the range
for num in range(lower, upper + 1):
    # Prime numbers must be greater than 1
    if num > 1:
        # Inner loop: Checks for factors from 2 up to num-1
        for i in range(2, num):
            if (num % i) == 0:
                # If a factor is found, it's not prime; break the inner loop
                break
        else:
            # If the inner loop finishes without finding a factor
            print(num)
```

```python
# Outer loop: Iterates through table numbers 1 to 20
for i in range(1, 21):
    print(f"--- Table of {i} ---")

    # Inner loop: Iterates through multipliers 1 to 10
    for j in range(1, 11):
        product = i * j
        print(f"{i} x {j} = {product}")

    # Prints a newline for better spacing between tables
    print()
```

```python
# 1. THE DEFINITION (Storing the instructions)
def hello():
    return "Hello, World!"

# 2. THE CALL (Executing the instructions)
print(hello())
```
